```python
import os
from datetime import datetime
import pandas as pd
```

```python
def read_multi_csv(path: str, df = None):
  for file in os.listdir(path):
      if not file.endswith(".csv"):
        continue
      src = os.path.join(path, file)
      temp = pd.read_csv(src, names=[*df.columns], header=None)
      if df is None:
        df = temp
      else:
        df = df.append(temp, ignore_index=True)
  return df
```

```python
def without_hurricane(df):
  return len(df[df.maximum_sustained_wind == 0].index)/len(df.index)
```

# LSTM

```python
from keras.layers import Input, Dense, LSTM, MaxPooling1D, Conv1D
from keras.models import Model, Sequential
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot
from math import sqrt

import pandas as pd
import numpy as np
```

```python
LAG = 12
DELTA = 5
```

```python
df = df.fillna(value=0)
```

```python
assert len(df) == len(df.groupby('fecha_corta').count())
```

```python
df.columns
```

```python
df.drop(columns=['cluster'], inplace=True)
```

```python
l = list(df.columns)
l.remove('fecha_corta')
```

```python
for i in l:
  df[i] = df[i].astype('float64')
df.info()
```

#### Quitar dÃ­as sin huracan

```python
#porcentaje de observaciones sin huracan
without_hurricane(df)
```

```python
PERCENTAGE = 0.6
df_balanced = df.drop(df[df.maximum_sustained_wind == 0].sample(int(len(df.index)*PERCENTAGE)).index)
print(without_hurricane(df_balanced))
df_balanced
```

```python
#porcentaje de observaciones sin huracan (Debe ser ahora 50%)
df = df_balanced
```

#### Normalizar tabla

```python
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
df[l] = scaler.fit_transform(df[l])
df.head()
```

```python
col_order = list(df.columns)
x_cols = sorted(list({col for col in df.columns}.difference(['fecha_corta', 'maximum_sustained_wind', 'day'])), key=lambda x: col_order.index(x))
```

```python
x_cols_t = []
for i in range(LAG,DELTA, -1):
  for x in x_cols:
    name = f't-{i}_{x}'
    df[name] = df[x].shift(i)
    x_cols_t.append(name)
df = df[LAG:]
df
```

```python
df.columns
```

```python
col_order = list(df.columns)
final_cols = []
final_cols.extend(sorted(list({col for col in df.columns}.difference(x_cols, *final_cols)), key=lambda x: col_order.index(x)))
df = df[final_cols]
df.head()
```

#### Crear modelo

```python
n_features = len(x_cols)*(LAG-DELTA)
shape = (n_features, 1)
shape
```

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=shape))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.summary()
```

```python
n_obs = len(df)
n_obs
```

```python
df.head()
```

```python
n_train = int(n_obs*0)
df_train, df_test = df[:n_train], df[n_train:]
df_train.info()
```

```python
print(without_hurricane(df_train))
print(without_hurricane(df_test))
```

```python
df_test.info()
```

```python
x_train = df_train[x_cols_t]
x_test = df_test[x_cols_t]
x_train.head()
```

```python
y_train = df_train["maximum_sustained_wind"]
y_test = df_test["maximum_sustained_wind"]
y_train.head()
```

```python
history = model.fit(x_train, y_train, epochs=50, batch_size=128, validation_data=(x_test, y_test), verbose=2, shuffle=False)
```

```python
y_predicted = model.predict(x_test)
```

```python
df_eval = df_test[['fecha_corta', 'maximum_sustained_wind']]
df_eval['prediction'] = y_predicted
```

```python
column = "prediction"
df_eval["prob"] = (df_eval[column] - df_eval[column].min()) / (df_eval[column].max() - df_eval[column].min())
```
